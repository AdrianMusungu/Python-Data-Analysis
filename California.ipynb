```python
#Assuming population_total to be our target value, fit the data onto a linear regression model and evaluate the performance of the model 

#here, we use the separator to avoid any Key Errors
mdf=pd.read_csv('mdf.csv', sep='\s*,\s*')
#here we define our target value 
r =mdf['area_total_km2'].values
s =mdf['population_total'].values
m =len(r)

from sklearn.linear_model import LinearRegression

#cannot use Rank 1 matrix in scikit learn
r = r.reshape((m, 1))

#creating model
reg = LinearRegression()

#Fitting training data
reg = reg.fit(r,s)

#Y- prediction
Y_pred = reg.predict(r)

#Calculating R2 Score

r2_score = reg.score(r,s)


print('An ideal r2 score is 1')
print('Ours is:')
print(r2_score)
print('')

```
